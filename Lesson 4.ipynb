#### Задание 2

```python
t = np.linspace(0, 10, 51)
print(t)
```

```python
f = np.cos(t)
print(f)
```

```python
bar = pd.Series(f, t)
bar.plot(kind = 'bar', title = 'График f(t)', xlabel = 'Значения t', ylabel = 'Значения f')
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.show()
```

#### *Задание 3

```python
x = np.linspace(-3, 3, 51)
x = x.astype(np.float32)
print(x)
```

```python
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2)
ax1, ax2, ax3, ax4 = ax.flatten()
ax1.plot(x, y1)
ax2.plot(x, y2)
ax3.plot(x, y3)
ax4.plot(x, y4)
```
